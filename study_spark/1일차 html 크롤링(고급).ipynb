```python
html="""
<!DOCTYPE html>
<html>
  <head>
     <meta charset='utf-8'>
     <title>Test BeautifulSoup</title>
  </head>
  <body>
     <p align="center"> text contents </p>
     <p align="right" class="myp">  text contents 2 </p>
     <p align="left" a="b">   text contents 3 </p>
     <img src="http://unico2013.dothome.co.kr/image/flower.jpg" width="500">
  </body>
</html> """


from bs4 import BeautifulSoup
bs = BeautifulSoup(html, 'html.parser')
```

```python
print(bs.find('p', align="center").text)
print(bs.find('p', class_="myp"))
print(bs.find('p', align="left"))
print("-------------------------------------")
print(bs.find('p', attrs={"align":"center"}))
print(bs.find('p', attrs={"align":"right", "class":"myp"}))
print(bs.find('p', attrs={"align":"left"}))

#text 공백제거
plist = bs.find('p', align="left")
print("["+plist.text+"]")
print("["+plist.get_text(strip=True)+"]")
print("["+plist.string.strip()+"]")
```

```python
#     <p align="left" a="b">   text contents 3 </p>
#text 공백제거
plist = bs.find('p', align="left")
print("["+plist.text+"]")
print("["+plist.get_text(strip=True)+"]")  # get_text 와 text 차이 보여준다. (공백을 제거함)
print("["+plist.string.strip()+"]")
```

```python
import requests
from bs4 import BeautifulSoup

header = {'User-Agent':'Mozila/5.0 (Winodws NT 6.3; Trident/7.0; rv:11.0) like Gecko'}
url =  'https://www.melon.com/chart/index.htm'
res = requests.get(url, headers=header)
res.text
```

```python
html=res.text
soup = BeautifulSoup(html,'html.parser')
soup
```

```python
song = 'ellipsis rank01'
singer= 'ellipsis rank02'

song_list = soup.find_all('div',{'class':song})
singer_list = soup.find_all('div',{'class':singer})

```

```python
#노래제목 추출
for song_item in song_list:
    print(song_item.find('a').text)
    
```

```python
# 가수 추출
for singer_item in singer_list:
    print(singer_item.find('a').text)
```

```python
s_list = []
sg_list= []

for song_item in song_list:
    s_list.append(song_item.find('a').text)
    
for singer_item in singer_list:
    sg_list.append(singer_item.find('a').text)
     
for i in range(100):
    print(f"{i+1}위 {s_list[i]}: {sg_list[i]}")
    
```
