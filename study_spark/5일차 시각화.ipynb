```python
st1=[100,110, 90, 105, 102, 85]
st2=[130,120, 91, 102, 12, 89]

x =  list(range(0,len(st1)))
x
```

```python
plt.title('Test Graph')
plt.plot(x, st1,color='red' ,label='st1')
plt.plot(x, st2, label='st2')
plt.xlabel('xxx')
plt.ylabel('yyy')
plt.legend(loc=1) #0~10
plt.grid(True)
```

```python
#  
#x    .
import numpy as np
x_index = np.arange(len(st1))

# x ,  width .
plt.title('bar Graph')
plt.bar(x_index -0.2, st1,width=0.2,color='green' ,label='')
plt.bar(x, st2,width=0.2,color='blue', label='')
plt.xlabel('X')
plt.ylabel('Y')

# plt.legend()
plt.legend(loc=1) #0~10
plt.grid(True)
plt.show()
```

```python
# 

import matplotlib.pyplot as plt
ratio =[34,20,15,31]
label = ['apple','grape','banana','melon']

plt.pie(x=ratio, labels=label)
plt.show()
```

```python
# 

import matplotlib.pyplot as plt
ratio =[34,20,15,31]
label = ['apple','grape','banana','melon']

plt.pie(x=ratio, labels=label, autopct="%.1f%%")
plt.show()
```

```python
explode=[0.01,0.20,0.01,0.01]

plt.pie(x=ratio, labels=label, autopct="%.1f%%",explode=explode)
plt.show()
```

```python
# Seaborn

import seaborn as sns
import pandas as pd


csData = pd.read_csv("C:\\Users\\daesi\\Downloads\\customerdata.csv")
csData.head(2)
```

```python
sns.countplot (data=csData, x='EMI', hue='CUSTTYPE')
```

```python
#
figure = plt.figure(figsize=(18,6))

ax1 = plt.subplot(1,2,1)

sns.distplot(csData.AVGPRICE, kde = False, ax=ax1 )

ax2 = plt.subplot(1,2,2)

sns.distplot(csData.AVGPRICE, hist = False, ax=ax2 )


```

```python
sns.barplot(x="EMI", y="AVGPRICE", data = csData, 
         hue = "CUSTTYPE", ci = "sd")
```
