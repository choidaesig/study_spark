```python
import requests
from bs4 import BeautifulSoup

url = 'https://tv.naver.com/r/'
res=requests.get(url)
res.text
```

```python
html=res.text
soup = BeautifulSoup(html,'html.parser')
soup
```

```python
topic='cds_info'
topic_list = soup.find_all('dl',{'class':topic})
```

```python
for topic_item in topic_list:
    print(topic_item.find('a').text)
```

```python
t_list=[]
for topic_item in topic_list:
    t_list.append(topic_item.find('a').text)

for i in range(100):
    print(f"{i+1}위 {t_list[i]}")
```
