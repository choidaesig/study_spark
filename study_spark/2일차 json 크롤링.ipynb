```python
url='http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=105'

savefile = 'c:\output'
```

```python
from bs4 import BeautifulSoup
import urllib.request as req
import io

req.urlretrieve(url, savefile)
```

```python
xml = open(savefile, 'r', encoding='utf-8').read()
xml
```

```python
soup = BeautifulSoup(xml, 'xml')
```

```python
#지역의 데이터 딕셔너리
info = {} #도시: 온도리스트[]

for location in soup.find_all('location'):
    loc=location.find('city').string
    min_w = location.find_all('tmn')
    max_w = location.find_all('tmx')
    
    weather = [a.string+"~"+b.string for a, b in zip(min_w, max_w)] # 3~15 묶어주는 코드
    
    if not loc in info:
        info[loc]=[]
    
    for data in weather:
        info[loc].append(data)
info
```

```python
with open('c:\output','wt',encoding='utf-8') as f:
     for loc in info.key():
        f.write(str(loc)+'\n')
        for item in info[loc]:
            f.write('\t'+str(item)+'\n')
```
