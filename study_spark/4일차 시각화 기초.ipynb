```python
import matplotlib.pylab as plt
```

```python
st1=[100,110, 90, 105, 102, 85]
st2=[130,120, 91, 102, 12, 89]

x =  list(range(0,len(st1)))
x
```

```python
plt.plot(x, st1)
```

```python
plt.plot(x, st2)
```

```python
plt.plot(x, st1)
plt.plot(x, st2)
```

```python
plt.title('Test Graph')
plt.plot(x, st1, label='st1')
plt.plot(x, st2, label='st2')
plt.xlabel('xxx')
plt.ylabel('yyy')
plt.legend(loc=10) #0~10
```

```python
plt.title('Test Graph')
plt.bar(x, st1, label='st1')
plt.bar(x, st2, label='st2')
plt.legend(loc=10) #0~10
```
