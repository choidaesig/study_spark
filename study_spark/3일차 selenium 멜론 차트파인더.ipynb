```python
#BS로 파싱하는 부분

from bs4 import BeautifulSoup

html = driver.page_source
soup= BeautifulSoup(html, 'html.parser')

#노래 제목 추출
data1 = soup.find_all('div',{'class':'ellipsis rank01'})

song_list=[]

for title in data1:
    song_list.append(title.find('a').get_text())

    
#가수  이름 가져오기 
data2 = soup.find_all('span', {'class': 'checkEllipsis'})

singer_list=[]

for title in data2:
    singer_list.append(title.find('a').get_text())
    
singer_list
```

```python
s_list = []
sg_list= []

for song_item in data1:
    s_list.append(song_item.find('a').text)
    
for singer_item in data2:
    sg_list.append(singer_item.find('a').text)
     
for i in range(100):
    print(f"{i+1}위 {s_list[i]}: {sg_list[i]}")
```
